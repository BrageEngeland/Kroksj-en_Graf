```python
class Solution(object):
    def twoSum(self, nums, target):
        output = []
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    output.append(i)
                    output.append(j)
                    return output
        return output
        
solution = Solution()
nums = [2, 7, 11, 15]
target = 9
print(solution.twoSum(nums, target))
```

```python
class Solution(object):
    def isPalindrome(self, x):
        if x < 0:
            return False
        
        x_str = str(x)
        return x_str == x_str[::-1]
    
solution = Solution()
print(solution.isPalindrome(131))  # Output: True
print(solution.isPalindrome(-131)) # Output: False
print(solution.isPalindrome(123))  # Output: False
```

```python
class Solution(object):
    def findMaxConsecutiveOnes(self, nums):
        current_count = 0
        max_count = 0
        for i in range(len(nums)):
            if nums[i] == 1:
                current_count += 1
                if current_count > max_count:
                    max_count = current_count
            else:
                current_count = 0
                       
        return max_count
```

```python
class Solution(object):
    def sortedSquares(self, nums):
        for i in range(len(nums)):
            nums[i] = nums[i]**2
        
        for j in range(len(nums)):
            for k in range(j+1, len(nums)):
                if(nums[j] > nums[k]):
                    nums[j], nums[k] = nums[k], nums[j]
        return nums

solution = Solution()
x = []
for i in range(-1000,1000):
    x.append(i)
print(solution.sortedSquares(x))
```

```python
class Solution(object):
    def duplicateZeros(self, arr):
        i = 0
        while i < len(arr):
            if(arr[i] == 0):
                arr.insert(i + 1, 0)
                arr.pop()
                i+=1
            i += 1
        return arr
            
print(Solution().duplicateZeros([1,0,2,3,0,4,5,0]))
```

```python
class Solution(object):
    def merge(self, nums1, m, nums2, n):
        overflydig = len(nums1) - m
        for i in range(0, overflydig):
            nums1.pop()
        overflydig2 = len(nums2) - n
        for i in range(len(nums2)):
            nums1.append(nums2[i])
        for j in range(len(nums1)):
            for k in range(j+1, len(nums1)):
                if(nums1[j] > nums1[k]):
                    nums1[j], nums1[k] = nums1[k], nums1[j]

        return nums1
            


solution = Solution()
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3
print(solution.merge(nums1, m, nums2, n))
```

```python
class Solution(object):
    def removeElement(self, nums, val):
        for i in range(len(nums) - 1, -1, -1):
            if nums[i] == val:
                del nums[i]
        return len(nums)
    
solution = Solution()
nums = [1,2,3,3,3,4,5,6,7]
print(solution.removeElement(nums,3))
```

```python
class Solution(object):
    def removeDuplicates(self, nums):
        for i in range(len(nums) - 1, -1, -1):
            for j in range(i - 1, -1, -1):
                if nums[i] == nums[j]:
                    del nums[j]
                    break
        return nums
solution = Solution()
nums = [1,1,1]
print(solution.removeDuplicates(nums))
```

```python
class Solution(object):
    def checkIfExist(self, arr):
        for i in range(len(arr)):
            for j in range(len(arr)):
                if(i != j and arr[i] == 2* arr[j]):
                    return True
        return False
        """
        :type arr: List[int]
        :rtype: bool
        """

solution = Solution()
nums = [7,1,14,11]
print(solution.checkIfExist(nums))
```
